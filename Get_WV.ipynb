```python
from nltk.parse.stanford import StanfordDependencyParser
eng_parser = StanfordDependencyParser(r"/Users/denhiroshi/Downloads/stanford-parser-full-2018-10-17/stanford-parser.jar",r"/Users/denhiroshi/Downloads/stanford-parser-full-2018-10-17/stanford-parser-3.9.2-models.jar")
infos = []
start = time.time()
sentences = []
for q in range(1,8):
    for temp in T[str(q)]:
        words = temp.split(" ")
        # 消除<e1></e1>,<e2></e2>合并实体(若中间有空格)
        el1 = 0
        er1 = 0        
        el2 = 0
        er2 = 0

        for i in range(len(words)):
            if words[i].find('<e1>') != -1:
                words[i] = words[i].replace('<e1>','').replace('</e1>','')
                el1 = i
                er1 = el1
            elif words[i].find('</e1>')  != -1:
                words[i] = words[i].replace('</e1>','')
                er1 = i
            elif words[i].find('<e2>')  != -1:
                words[i] = words[i].replace('<e2>','').replace('</e2>','')
                el2 = i
                er2 = el2
            elif words[i].find('</e2>')  != -1:
                words[i] = words[i].replace('</e2>','')
                er2 = i
            else:
                continue

        # 清洗后已经分词了的句子
        new_words = []
        j = 0
        while(j < len(words)):
            if j == el1:
                temp_words = ""
                while j <= er1:
                    temp_words = temp_words+" "+words[j]
                    j+=1
                new_words.append(temp_words.replace("  "," ").lstrip().rstrip())
            elif j == el2:
                temp_words = ""
                while j <= er2:
                    temp_words = temp_words+" "+words[j]
                    j+=1
                new_words.append(temp_words.replace("  "," ").lstrip().rstrip())
            else:
                new_words.append(words[j].replace("  "," ").lstrip().rstrip())
                j+=1

        # 使用的当前句子
        use = new_words
        use_map = {}
        for i in range(len(use)):
            use_map[use[i]] = i
            
        sentences.append(use)
        res = list(eng_parser.parse(use))
        info = []
        for row in res[0].triples():
            info.append(row)
        infos.append(info)

        # 遍历所有句子，找出当前句子在其他句子的共现矩阵
        word_mat_T = []
        for zz in range(1,8):
            for temp_T in T[str(zz)]:
                words = temp_T.split(" ")
                for i in range(len(words)):
                    if words[i].find('<e1>') != -1:
                        words[i] = words[i].replace('<e1>','').replace('</e1>','')
                        el1 = i
                        er1 = el1
                    elif words[i].find('</e1>')  != -1:
                        words[i] = words[i].replace('</e1>','')
                        er1 = i
                    elif words[i].find('<e2>')  != -1:
                        words[i] = words[i].replace('<e2>','').replace('</e2>','')
                        el2 = i
                        er2 = el2
                    elif words[i].find('</e2>')  != -1:
                        words[i] = words[i].replace('</e2>','')
                        er2 = i
                    else:
                        continue

                # 清洗后已经分词了的句子
                new_words = []
                j = 0
                while(j < len(words)):
                    if j == el1:
                        temp_words = ""
                        while j <= er1:
                            temp_words = temp_words+" "+words[j]
                            j+=1
                        new_words.append(temp_words)
                    elif j == el2:
                        temp_words = ""
                        while j <= er2:
                            temp_words = temp_words+" "+words[j]
                            j+=1
                        new_words.append(temp_words)
                    else:
                        new_words.append(words[j])
                        j+=1

                word_mat = np.zeros((len(use),len(use)))
                for u in range(len(new_words)):
                    if new_words[u] in use:
                        # 获取原句子中当前词的窗口
                        use_window = []
                        index = use.index(new_words[u])
                        for z in range(1,biasl+1):
                            flag = index-z
                            if flag < 0:
                                break
                            else:
                                use_window.append(use[flag])

                        for z in range(1,biasr):
                            flag = index+z
                            if flag >= len(use):
                                break
                            else:
                                use_window.append(use[flag])

                        # 获取其他句子中当前词的窗口(有此单词且窗口中有原窗口词则加)
                        for z in range(1,biasl+1):
                            flag = u - z
                            if flag < 0:
                                break
                            else:
                                if new_words[flag] in use_window:
                                    x = use_map[new_words[u]]
                                    y = use_map[new_words[flag]]
                                    word_mat[x][y]+=1

                        for z in range(1,biasr):
                            flag = u + z
                            if flag >= len(new_words):
                                break
                            else:
                                if new_words[flag] in use_window:
                                    x = use_map[new_words[u]]
                                    y = use_map[new_words[flag]]
                                    word_mat[x][y]+=1
                word_mat_T.append(word_mat)
        Tensor.append(np.array(word_mat_T))
print(time.time() - start)
```

```python
import torch
from torch.autograd import Function
from torch.autograd import Variable
import tensorly as tl
from tensorly.decomposition import parafac
from tensorly.decomposition import tucker
from tqdm import tqdm
import time
import math
import logging
lr = 0.05
n_epoches = 100
tl.set_backend('pytorch')
help(tl.set_backend)
```

```python
def get_loss(data,Y,weight):
    torch.set_grad_enabled(True)
    a,b,c = parafac(data,Y.shape[0])
    b.requires_grad_()
    c.requires_grad_()
    loss_fn = torch.nn.SmoothL1Loss(reduce=False, size_average=False) 
    output_linear = torch.mm(torch.mm(b,weight),c).double()
    loss = loss_fn(output_linear,Y)
    return loss.mean()

def asMinutes(s):
    m = math.floor(s / 60)
    s -= m * 60
    return "%dm %ds" % (m, s)

def timeSince(start, percent):
    now = time.time()
    s = now - start
    r = s / (percent)
    return "%s (- %s)" % (asMinutes(s), asMinutes(r - s))
```

```python
def train(data, Y, weight, num_epochs, lr, Embeddings):
    opt_Adam = torch.optim.Adam((data,weight), lr=lr, betas=(0.9, 0.99))
    opt_Adam.zero_grad()
    start = time.time()
    for epoch in tqdm(range(1,n_epoches+1)):
        logging.info("start epoch %i", epoch)
        aver_loss = 0
        opt_Adam.zero_grad()
        loss = get_loss(data, Y, weight)
        loss.backward()
        opt_Adam.step()
        logging.info("%s (%d %d%%) %.4f" % (timeSince(start, epoch / n_epoches),
                                         epoch, epoch / n_epoches * 100, loss.item()))
        if epoch == n_epoches:
            a,b,c = parafac(data,Y.shape[0])
            Embeddings.append(a)
            logging.info("get a Sentence Embeddings")
```

```python
def get_weight(x):
    if x < x_max:
        return (x*1.0/x_max)**(0.75)
    else:
        return 1.0
```

```python
def get_Sum_J(word_vector,b_i,b_j):
    Sum_J = 0
    for i in range(word_vector.shape[0]):
        for j in range(word_vector.shape[1]):
            Sum_J += get_weight(word_vector[i][j]) * (np.dot(word_vector[i].T,word_vector[j]) + b_i + b_j - np.log(word_vector[i][j]+0.001))**2
    return Sum_J
```
